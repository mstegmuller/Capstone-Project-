I had to convert the inflation rate value, with multiplication, to see if there was any correlation between the total costs of disasters. Since inflation has steadily risen, the costs of disasters are a factor in inflation, but not the main reason for the rise.

```python
plt.figure(figsize=(20,20))
x = Inflation_rate_vs_total_cost.year_only
y1 = Inflation_rate_vs_total_cost['Inflation rate value']
y2 = Inflation_rate_vs_total_cost['Total CPI-Adjusted Cost (Millions of Dollars)']

plt.plot(x, y1, label ='Cost in Millions')
plt.plot(x, y2, '-.', label ='Inflation Rate multiplied by 1000')

plt.xlabel("Year")
plt.ylabel("cost in millions : inflation rate (divide by 1000)")
plt.legend()
plt.title('Disaster costs vs inflation rate')
plt.show()
```
